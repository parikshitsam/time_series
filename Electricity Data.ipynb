```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.tsa.stattools as tsa
%matplotlib notebook
```

```python
path = r"C:\Users\parik\Documents\Data sets\elecprices.csv"
df = pd.read_csv(path)
df.set_index("Year",inplace=True)
df.head()
```

```python
# df = df[90-48:]
df.plot()
```

```python
df['lnprice'] = np.log(df)
```

```python
plt.figure()
df['diff'] = df['lnprice'].diff(periods=1)
df["diff"].plot()
```

```python
adstat, pvalue, used_lag,*rest = tsa.adfuller(df['diff'][1:])
print("adfstat: ",adstat, "     ", "Pvalue: ", pvalue, "     ", "Num Lags: ",used_lag)
```
